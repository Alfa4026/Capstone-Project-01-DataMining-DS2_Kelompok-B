## Mengoreksi tipe data

```python
df.info()
```

Kami akan mengubah beberapa tipe data pada beberapa variabel yang tipenya tidak sesuai. Namun sebelum itu perlu diperiksa nilai entri-entri di setiap variabel, sebab bisa saja responden mengisi nilai yang tidak sesuai pada variabel yang diminta. Akibatnya akan menimbulkan error jika kami langsung mengubah tipe data. Oleh karena itu sebelum mengubah tipe data, kami akan memeriksa entri-entri di setiap variabel. Akan tetapi beberapa variabel yang bisa langsung kami ubah saat ini karena tidak akan menimbulkan error.

```python
def cekdata(df):
  list_kolom = list(df.columns)
  for i in list_kolom:
    print("\n"f'variabel', i)
    print(df[i].unique())

cekdata(df)
```

```python
# Ubah tipe data kolom
df['Waktu Mendapatkan Pekerjaan:'] = df['Waktu Mendapatkan Pekerjaan:'].astype('category')
df['Status saat ini:'] = df['Status saat ini:'].astype('category')
df['Tahun Wirausaha'] = pd.to_numeric(df['Tahun Wirausaha'], errors='coerce').astype('Int64')
df['Program Studi'] = df['Program Studi'].astype('category')
# Daftar bulan valid
bulan_valid = ['Januari', 'Februari', 'Maret', 'April', 'Mei','Juni', 'Juli', 'Agustus',
               'September', 'Oktober', 'November', 'Desember']

# Membersihkan spasi ekstra dan menerapkan filter bulan valid
df['Bulan Wisuda'] = (
    df['Bulan Wisuda']
    .str.strip()  # Menghilangkan spasi di awal/akhir
    .where(lambda x: x.isin(bulan_valid), pd.NA)  # Mempertahankan bulan valid, sisanya jadi pd.NA
)
df['Bulan Wisuda'] = df['Bulan Wisuda'].astype('category')
# Mengubah kolom Kesesuaian Kurikulum dan Kesesuaian Bidang Ilmu menjadi kategori ordinal
kategori_ordinal = pd.api.types.CategoricalDtype(categories=[1, 2, 3, 4, 5], ordered=True)
df['Kesesuaian Kurikulum:'] = df['Kesesuaian Kurikulum:'].astype(kategori_ordinal)
df['Kesesuaian Bidang Ilmu:'] = df['Kesesuaian Bidang Ilmu:'].astype(kategori_ordinal)

# Tampilkan informasi data untuk memastikan perubahan berhasil
print(df.info())
```

## Visualisasi terkait Lama Studi berdasarkan Program Studi FST UIN Jakarta.

```python
# Menghitung modus untuk setiap program studi
mode_data = df.groupby('Program Studi')['Lama Studi di UIN'].agg(pd.Series.mode).reset_index()

plt.figure(figsize=(12, 6))

# Membuat bar plot berdasarkan modus
barplot = sns.barplot(x='Program Studi', y='Lama Studi di UIN', data=mode_data, ci=None, palette="Paired")

# Menambahkan nilai di atas setiap bar
for p in barplot.patches:
    barplot.annotate(f'{p.get_height():.2f}',
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='bottom',
                     fontsize=10)

plt.title('Lama Studi berdasarkan Program Studi')
plt.xticks(rotation=45)
plt.ylabel('Lama Studi (Tahun)')
plt.show()
```

**Insight**: Dari visualisasi data di atas dapat dilihat bahwa 4 program studi teratas dalam hal lama studi atau lulusannya memerlukan waktu yang lama untuk menyelesaikan studi di prodi tersebut yaitu prodi agribisnis S1 kebanyakan lulusannya menghabiskan waktu 7 tahun disusul prodi biologi sekitar 5 tahun. Lalu prodi teknik informatika sekitar 5 tahun begitu juga prodi teknik pertambangan sekitar 5 tahunan.

**Saran**: Pihak fakultas kedepannya harus melakukan analisa lebih lanjut terkait apa saja faktor-faktor yang membuat mahasiswa lama menyelesaikan studi, karena salah satu faktor yang dapat meningkatkan akreditasi suatu program studi yaitu dilihat dari lulus tepat waktu atau tidaknya seorang mahasiswa/i. Dengan demikian akreditasi tiap-tiap program studi bisa semakin membaik dari tahun ke tahun.
